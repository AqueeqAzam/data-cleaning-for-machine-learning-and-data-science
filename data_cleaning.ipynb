# `Alternative Method of missing value`

```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/AqueeqAzam/DS-and-ML-datasets/main/test.csv')
# df.head()
df.head(5)

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer=imputer.fit(df['Age'])
df[:, 1:3] = imputer.transform(df['Age'])
df

```

# `One Hot Encoding`

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/AqueeqAzam/DS-and-ML-datasets/main/test.csv')
df.head(3)

df['Salery'].unique()

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# fit the model of encoding
le.fit(df['Salery'])

# display encoding in array form
le.transform(df['Salery'])

df['encoder name'] = le.fit_transform(df['Salery'])
df.head(6)
```

# `Using Dictionary`

# `Ordinal encoding`

```python
import pandas as pd
df = pd.DataFrame({'Size': ['s', 'm', 'l', 'xl', 's', 's', 'l', 'm']})
df.head(4)

from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()

# learns the categories and creates a mapping.
oe.fit(df[['Size']])

# convert the data
oe.transform(df[['Size']])

# display the mapping
oe.categories_

df['encoder_size'] = oe.fit_transform(df[['Size']])
df
```

# `Binarization`

binarization code refers to code that converts data from its original format (usually continuous or categorical) into a binary format, where each value is either 0 or 1.

```python
from sklearn.preprocessing import Binarizer

# Create a Binarizer object with a threshold of 5
binarizer = Binarizer(threshold=5)

# Binarize a sample data array
data = [3, 8, 1, 10]
binary_data = binarizer.transform([data])  # binary_data will be [0, 1, 0, 1]
binary_data
```

# `Standerization`

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# This line reads data from a CSV file located at the specified URL and stores it in a pandas DataFrame with variable named df.
df = pd.read_csv("https://raw.githubusercontent.com/AqueeqAzam/DS-and-ML-datasets/main/disbetes.csv")
df.head(4)

#  Assuming df is a DataFrame containing your data, this selects the column named "Insulin".
# Inside the .fillna method, we calculate the mean of the "Insulin" column using df['Insulin'].mean().
#  inplace=True argument ensures this modification happens directly on the original DataFrame df.

df['Insulin'].fillna(df['Insulin'].mean(), inplace=True)
df
sns.displot(df['Insulin'])
plt.show()

# Scalinfg the data
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(df[['Insulin']])
ss.transform(df[['Insulin']])

df['scaled_insulin'] = ss.fit_transform(df[['Insulin']])
df.head(5)
```

# `Handle Duplicate Data`
